## Data Scrpaing from clubs page

```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
no = 0
clubs_info = []
url = "https://fstt.ac.ma/Portail2023/clubs/"

result = requests.get(url)
soup = BeautifulSoup(result.content, "lxml")

all_clubs_container = soup.find("div", {"class": "elementor-element elementor-element-78ced8b elementor-widget elementor-widget-theme-post-content"})

all_clubs = all_clubs_container.find_all(lambda tag: tag.name == 'div' and 'class' in tag.attrs and 'ae-link-yes' in tag['class'])

for club in all_clubs :
    club_name = club.contents[1].find("h3").string.strip()
    club_link = club.attrs["data-wts-url"]

    result2 = requests.get(club_link)
    soup2 = BeautifulSoup(result2.content, "lxml")

    club_info = soup2.find("div", {"class": "elementor-widget-theme-post-content"}).text.strip()

    no += 1

    clubs_info.append({
        "No": no,
        "club_name": club_name,
        "departement_info": club_info,
        "club_link" : club_link
    })
```

```python
import pandas as pd

pd.DataFrame(clubs_info)
```

```python
columns = clubs_info[0].keys()

with open("fstt-clubs-info.csv", "w", encoding="utf-8-sig", newline="") as csv_file:
    dict_writer = csv.DictWriter(csv_file, columns)
    dict_writer.writeheader()
    dict_writer.writerows(clubs_info)

    print("*"*50, "\n file created \n", "*"*50)
```
