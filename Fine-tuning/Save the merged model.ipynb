```python
from peft import PeftConfig
```

```python
adaptor = "aymanboufarhi/chat-bot-fstt"
config = PeftConfig.from_pretrained(adaptor)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from peft import PeftModel
import torch



model = AutoModelForCausalLM.from_pretrained(
        config.base_model_name_or_path,
        return_dict=True,
        low_cpu_mem_usage=True,
        torch_dtype=torch.float16,
        device_map="auto",
        trust_remote_code=True,
)
```

```python
model = PeftModel.from_pretrained(model, adaptor)
```

```python
modelmerged = model.merge_and_unload()
```

```python
tokenizer = AutoTokenizer.from_pretrained("google/gemma-2b-it", trust_remote_code=True)
```

```python
tokenizer.push_to_hub("aymanboufarhi/2B-chat-bot-fstt", use_auth_token=True)
```

```python
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"
```

```python
prompt = '''<|system|>FSTT c'est la Faculté des Sciences et Techniques de Tanger 
<|user|> addresse de FSTT ?
<|assistant|>'''
    
inputs = tokenizer(prompt, return_tensors='pt', padding=True, truncation=True).to("cuda")

outputs = model.generate(**inputs, max_length=500, num_return_sequences=1)

text = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(text)
```
